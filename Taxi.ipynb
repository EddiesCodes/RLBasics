```python
import gym
import numpy as np
```

```python
env = gym.make("Taxi-v3")
env.render()
```

```python
action_space = env.action_space.n 
observation_space = env.observation_space.n
action_space, observation_space
```

```python
Q = np.zeros((observation_space, action_space))
Q
```

```python
def epsilon_greedy(Q, eps, state):
    if (1-eps) > np.random.uniform():
        return np.argmax(Q[state])
    else:
        return env.action_space.sample()
```

```python
def TaxiQ(Q, eps, env, num_episodes, alpha, gamma, render):
    
    for i in np.arange(num_episodes):
        eps *= 0.95
        state = env.reset()
        while(True):
            action = epsilon_greedy(Q, eps, state)
            next_state, reward, done, _ = env.step(action)
            if(render == True):
                env.render()
            Q[state][action] += alpha*(reward+gamma*np.max(Q[next_state])-Q[state][action])
            if done == True:
                break
            state = next_state
    return Q
    
```

```python
Q = TaxiQ(Q, 1, env, 1000, 0.1, 0.99, False)
```

```python
import time
rewards = []
frames = []
for episode in range(1000):
    state = env.reset()
    step = 0
    done = False
    total_rewards = 0
    print("****************************************************")
    print("EPISODE ", episode)
    while(True):
        env.render()     
        # Take the action (index) that have the maximum expected future reward given that state
        action = np.argmax(Q[state])
        new_state, reward, done, _ = env.step(action)
        total_rewards += reward
        
        if done:
            rewards.append(total_rewards)
            #print ("Score", total_rewards)
            break
        state = new_state
env.close()
print ("Score over time: " +  str(sum(rewards)/1000))
```
